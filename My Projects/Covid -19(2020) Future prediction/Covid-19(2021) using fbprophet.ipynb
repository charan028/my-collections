# Prediction of Confirmed

```python
confirmed.columns=['ds','y']
```

```python
confirmed['ds']=pd.to_datetime(confirmed['ds'])
```

```python
confirmed
```

```python
model=Prophet(interval_width=0.95)# by default 80, confidence in stats
model.fit(confirmed)
```

```python
#training completed
# predicting future 
```

```python
future=model.make_future_dataframe(periods=100)
future
```

```python
forecast_confirmed=model.predict(future)
```

```python
forecast_confirmed
```

```python
forecast_confirmed[['ds','yhat','yhat_lower','yhat_upper']]
```

```python
forecast_confirmed[['ds','yhat','yhat_lower','yhat_upper']].tail(100)
# yhat-cases in future(based on pattern,data),yhat_lower(lowerlimit),yhat_upper(upperlimit)
```

```python
confirmed_forecast_plot=model.plot(forecast_confirmed)
```

```python
confirmed_forecast_plot_component=model.plot_components(forecast_confirmed)
```

```python
# doesn't make sense of weekly basis
```
